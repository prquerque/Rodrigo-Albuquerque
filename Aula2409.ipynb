```python
# questão 04 (classe)

class Produto:
    def __init__(self, id_produtos, nome, preco):
        self.id_produtos = id_produtos
        self._nome = ""  # Atributo protegido para nome
        self._preco = 0.0  # Atributo protegido para preço

        # Usando os setters para validar na inicialização
        self.set_nome(nome)
        self.set_preco(preco)

    # Getter e Setter para NOME
    def get_nome(self):
        return self._nome

    def set_nome(self, valor):
        if not valor or not valor.strip():
            raise ValueError("Erro: O nome não pode ser vazio!")
        self._nome = valor.strip()

    # Getter e Setter para PREÇO
    def get_preco(self):
        return self._preco

    def set_preco(self, valor):
        if valor <= 0:
            raise ValueError("Erro: O preço deve ser maior que zero!")
        self._preco = valor

    def __repr__(self):
        return f"Produto(id_produtos={self.id_produtos}, nome='{self._nome}', preco={self._preco})"


# Criando uma lista com 3 produtos
def criar_lista_produtos():
    produtos = []

    try:
        # Produto 1
        produto1 = Produto(1, "Notebook Gamer", 3500.00)
        produtos.append(produto1)

        # Produto 2
        produto2 = Produto(2, "Mouse RGB", 150.00)
        produtos.append(produto2)

        # Produto 3
        produto3 = Produto(3, "Teclado Mecânico", 450.00)
        produtos.append(produto3)

    except ValueError as e:
        print(f"Erro ao criar produto: {e}")

    return produtos


# Testando a lista de produtos
if __name__ == "__main__":
    # Criando a lista de produtos
    lista_produtos = criar_lista_produtos()

    # Exibindo todos os produtos
    print("=== LISTA DE PRODUTOS ===")
    for produto in lista_produtos:
        print(produto)

    print("\n=== TESTANDO GETTERS ===")
    # Testando os getters
    for produto in lista_produtos:
        print(f"ID: {produto.id_produtos}")
        print(f"Nome: {produto.get_nome()}")
        print(f"Preço: R$ {produto.get_preco():.2f}")
        print("-" * 30)

    print("\n=== TESTANDO SETTERS ===")
    # Testando os setters com valores válidos
    try:
        produto = lista_produtos[0]
        produto.set_nome("Notebook Gamer Pro")
        produto.set_preco(3800.00)
        print(f"Produto atualizado: {produto}")
    except ValueError as e:
        print(e)

    # Testando setters com valores inválidos
    print("\n=== TESTANDO VALIDAÇÕES ===")

    # Teste com preço inválido
    try:
        lista_produtos[1].set_preco(-50.00)
    except ValueError as e:
        print(f"Erro no preço: {e}")

    # Teste com nome vazio
    try:
        lista_produtos[2].set_nome("")
    except ValueError as e:
        print(f"Erro no nome: {e}")

    # Exibindo lista final
    print("\n=== LISTA FINAL DE PRODUTOS ===")
    for i, produto in enumerate(lista_produtos, 1):
        print(f"{i}. {produto}")
```

```python
# questão 05 (classe)

class Cliente:
    def __init__(self, id_cliente, nome, email):
        self._id_cliente = None  # Atributo privado
        self._nome = ""
        self._email = ""

        # Usando os setters para validar na inicialização
        self.set_id_cliente(id_cliente)  # Embora seja somente leitura, validamos na criação
        self.set_nome(nome)
        self.set_email(email)

    # Getter e Setter para ID_CLIENTE (somente leitura após criação)
    def get_id_cliente(self):
        return self._id_cliente

    def set_id_cliente(self, valor):
        if self._id_cliente is not None:
            raise AttributeError("Erro: id_cliente é somente leitura e não pode ser alterado!")

        if not valor or valor <= 0:
            raise ValueError("Erro: id_cliente deve ser um número maior que zero!")

        self._id_cliente = valor

    # Getter e Setter para NOME
    def get_nome(self):
        return self._nome

    def set_nome(self, valor):
        if not valor or not valor.strip():
            raise ValueError("Erro: O nome não pode ser vazio!")
        self._nome = valor.strip()

    # Getter e Setter para EMAIL
    def get_email(self):
        return self._email

    def set_email(self, valor):
        if not valor or not valor.strip():
            raise ValueError("Erro: O email não pode ser vazio!")

        if "@" not in valor:
            raise ValueError("Erro: O email deve conter o caractere '@'!")

        self._email = valor.strip()

    def __repr__(self):
        return f"Cliente(id_cliente={self._id_cliente}, nome='{self._nome}', email='{self._email}')"


# Função para criar uma lista de clientes de exemplo
def criar_lista_clientes():
    clientes = []

    try:
        # Cliente 1
        cliente1 = Cliente(1, "João Silva", "joao.silva@email.com")
        clientes.append(cliente1)

        # Cliente 2
        cliente2 = Cliente(2, "Maria Santos", "maria.santos@empresa.com")
        clientes.append(cliente2)

        # Cliente 3
        cliente3 = Cliente(3, "Pedro Oliveira", "pedro.oliveira@site.com")
        clientes.append(cliente3)

    except (ValueError, AttributeError) as e:
        print(f"Erro ao criar cliente: {e}")

    return clientes


# Testando a classe Cliente
if __name__ == "__main__":
    # Criando a lista de clientes
    lista_clientes = criar_lista_clientes()

    # Exibindo todos os clientes
    print("=== LISTA DE CLIENTES ===")
    for cliente in lista_clientes:
        print(cliente)

    print("\n=== TESTANDO GETTERS ===")
    # Testando os getters
    for cliente in lista_clientes:
        print(f"ID: {cliente.get_id_cliente()}")
        print(f"Nome: {cliente.get_nome()}")
        print(f"Email: {cliente.get_email()}")
        print("-" * 40)

    print("\n=== TESTANDO SETTERS (NOME E EMAIL) ===")
    # Testando os setters com valores válidos
    try:
        cliente = lista_clientes[0]
        cliente.set_nome("João da Silva")
        cliente.set_email("joao.novo@email.com")
        print(f"Cliente atualizado: {cliente}")
    except ValueError as e:
        print(e)

    print("\n=== TESTANDO VALIDAÇÕES ===")

    # Teste com nome vazio
    try:
        lista_clientes[1].set_nome("")
    except ValueError as e:
        print(f"Erro no nome: {e}")

    # Teste com email sem @
    try:
        lista_clientes[2].set_email("emailinvalido.com")
    except ValueError as e:
        print(f"Erro no email: {e}")

    # Teste com email vazio
    try:
        lista_clientes[2].set_email("")
    except ValueError as e:
        print(f"Erro no email: {e}")

    print("\n=== TESTANDO ID_CLIENTE (SOMENTE LEITURA) ===")
    # Tentativa de alterar id_cliente (deve falhar)
    try:
        lista_clientes[0].set_id_cliente(999)
    except AttributeError as e:
        print(f"Erro: {e}")

    # Exibindo lista final
    print("\n=== LISTA FINAL DE CLIENTES ===")
    for i, cliente in enumerate(lista_clientes, 1):
        print(f"{i}. {cliente}")
```

```python
# questão 06 (classe)

from abc import ABC, abstractmethod

class Pagamento(ABC):
    def __init__(self, valor):
        self.valor = valor

    @abstractmethod
    def processar_pagamento(self):
        pass

class PagamentoPix(Pagamento):
    def processar_pagamento(self):
        return f"Pagamento de R${self.valor:.2f} processado via PIX"

class PagamentoDinheiro(Pagamento):
    def processar_pagamento(self):
        return f"Pagamento de R${self.valor:.2f} processado em DINHEIRO"
```
